```python
import json
import random
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

DATAFILE = "reviews_10k.json"

def parse_reviews(filename):
    with open(DATAFILE, 'r') as fi:
        reviews = []
        for line in fi:
            try:
                reviews.append(json.loads(line))
            except:
                continue
        return reviews
```

```python
    sns.set_style('darkgrid')
    reviews = parse_reviews(DATAFILE) # :dict
    datetimes = [datetime.strptime(review['date'], "%Y-%m-%d")
                 for review in reviews]
    # just look at 2016 data
    epochs = [dt.timestamp() for dt in datetimes if dt.year == 2016]
    print(len(epochs))
    # convert the epoch format to matplotlib date format
    mpl_data = mdates.epoch2num(epochs)
    # plot it
    fig, ax = plt.subplots(1,1)
    ax.hist(mpl_data, color='lightblue')

    locator = mdates.AutoDateLocator()
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(locator))
    # ax.xaxis.set_major_locator(mdates.YearLocator())
    # ax.xaxis.set_major_formatter(mdates.DateFormatter('%m.%d.%y'))
    plt.show()
```

```python
def extract_month(datetimes, mon):
    return_list = []
    for dt in datetimes:
        if (dt.month == mon):
            return_list.append(dt)
    return len(return_list)
```

```python
count = extract_month(datetimes, 3)
#script to get count of reviews by month
review_count = []
for i in range(1, 13):
    review_count.append(extract_month(datetimes, i))
print(review_count)
```

```python
def month_count(month_data):
    return_list = []
    for i in range(0, 12):
        lower_bound = 4 * i
        upper_bound = (4 * i) + 4
        slic = month_data[lower_bound:upper_bound].values.tolist()
        count_list = []
        for lst in slic: 
            count_list.append(lst[0])
        return_list.append(sum(count_list))
    return return_list

month_counts = month_count(total_patients)
print(month_counts)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
regr = linear_model.LinearRegression()
#X-values is number of reservations, Y-values is the month_counts 
regr.fit(np.array(review_count).reshape(-1, 1), np.array(month_counts).reshape(-1, 1))
predictions = regr.predict(np.array(review_count).reshape(-1, 1))
```

```python
print("Mean squared error: %.2f"% mean_squared_error(np.array(review_count).reshape(-1, 1), predictions))
```

```python
print('Variance score: %.2f' % r2_score(np.array(review_count).reshape(-1, 1), predictions))
```

```python
#plt.scatter(np.array(review_count).reshape(-1, 1), predictions,  color='black')
plt.plot(np.array(review_count).reshape(-1, 1), predictions, color='blue', linewidth=3)
plt.scatter(np.array(review_count).reshape(-1, 1), np.array(month_counts).reshape(-1, 1))
```

```python
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
scaler_y = StandardScaler()
```

```python
preprocessed_x = scaler_x.fit(np.array(review_count).reshape(-1, 1))
preprocessed_y = scaler_y.fit(np.array(month_counts).reshape(-1, 1))
```

```python
print('Variance score: %.2f' % r2_score(preprocessed_y.transform(np.array(month_counts).reshape(-1, 1)).reshape(-1, 1), predictions_1))
print("Mean squared error: %.2f"% mean_squared_error(preprocessed_x.transform(np.array(review_count).reshape(-1, 1)).reshape(-1, 1), predictions_1))
```
